```r
getParticipantCode <- function(fullBarcode) {
  tcga.barcode.pattern <- '(TCGA-[A-Z0-9a-z]{2}-[a-zA-Z0-9]{4}).*'
  return(gsub(tcga.barcode.pattern, '\\1', fullBarcode))
}
```

```r
BLCAmut['Patient_Barcode'] <- lapply(BLCAmut['Tumor_Sample_Barcode'], getParticipantCode )
```

```r
(blca_pat <- unlist(unique(BLCAmut['Patient_Barcode']),use.names = FALSE))
```
